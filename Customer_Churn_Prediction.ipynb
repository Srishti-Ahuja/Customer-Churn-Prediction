```
def remove_na(df):
  mean_price_p1_var=np.mean(df.price_p1_var)
  mean_price_p2_var=np.mean(df.price_p2_var)
  mean_price_p3_var=np.mean(df.price_p3_var)
  mean_price_p1_fix=np.mean(df.price_p1_fix)
  mean_price_p2_fix=np.mean(df.price_p2_fix)
  mean_price_p3_fix=np.mean(df.price_p3_fix)
  df['price_p1_var']=df['price_p1_var'].fillna(mean_price_p1_var)
  df['price_p2_var']=df['price_p2_var'].fillna(mean_price_p2_var)
  df['price_p3_var']=df['price_p3_var'].fillna(mean_price_p3_var)
  df['price_p1_fix']=df['price_p1_fix'].fillna(mean_price_p1_fix)
  df['price_p2_fix']=df['price_p2_fix'].fillna(mean_price_p2_fix)
  df['price_p3_fix']=df['price_p3_fix'].fillna(mean_price_p3_fix)
  return df
```
